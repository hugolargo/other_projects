# cleaning artists column:

```python
df['artists'] = df['artists'].str.lstrip("['")
df['artists'] = df['artists'].str.replace(r"\']", '')
```

```python
df.head()
```

```python
df['artists'] = df['artists'].str.replace(r'\"]', '')
```

```python
df[df['artists'].str.contains("'")]
```

```python
df['artists'] = df['artists'].map(lambda x: x.lstrip("'").rstrip("'"))
```

```python
df['artists'] = df['artists'].str.replace(r'\', "', ", ")
```

```python
df['artists'] = df['artists'].str.replace(r'\', "', ", ")
```

```python
df[df['artists'].str.contains("'")]
```

```python
df['artists'] = df['artists'].map(lambda x: x.lstrip('"').rstrip('"'))
```

```python
df['duration'] = df['duration']/60000
```

```python
df.rename(columns={"duration": "duration_minutes", "name": "song_title"}, inplace=True)
```

# adding ISO-3 country codes for later map-plotting:

```python
df_countries = df[df['country']!='global']
df_countries
```

```python
from iso3166 import countries

country_names =[]
for country in df_countries['country']:
        c_name = countries.get(country)
        country_names.append(c_name[2])

df_countries['iso_alpha'] = country_names
```

```python
df_countries.head()
```

```python
# save this datafram to csv:
```

```python
df_countries.to_csv('countries_data_with_iso.csv')
```
