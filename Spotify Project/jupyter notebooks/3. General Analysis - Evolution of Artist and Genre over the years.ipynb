```python
import pandas as pd
```

```python
df = pd.read_csv('../data/global_spotify_charts.csv', index_col = [0])
df
```

# DRAKE over the years

```python
df_drake = df[df['artists'].str.contains('Drake')]
df_drake
```

```python
df_drake_by_year = df_drake.groupby(['year']).agg('sum').sort_values('year', ascending=True)
df_drake_by_year
```

```python
df_drake_by_year.reset_index(inplace=True)
```

```python
df_drake_by_year.drop(columns=['position', 'duration_minutes', 'explicit', 'month'], inplace=True)
df_drake_by_year
```

```python
df_drake_by_year
```

```python
df_drake_by_year.to_csv('drake_by_year.csv')
```

# Genres over the years

```python
df.head()
```

```python
df_pop = df[df['artist_genres'].str.contains("'pop'")]
df_pop
```

```python
df_pop_year = df_pop.groupby(['year']).agg('sum').sort_values('year', ascending=True)
df_pop_year
```

```python
genres = ['pop', 'metal', 'rock', 'punk', 'dance', 'electronic', 'rap', 'r&b', 'alternative', 'indie', 'jazz', 'classical']
df_all_genres = pd.DataFrame()

for genre in genres:
    df_temp = df[df['artist_genres'].str.contains(f"'{genre}'")]
    df_temp = df_temp.groupby(['year']).agg('sum').sort_values('year', ascending=True)
    df_temp.reset_index(inplace=True)
    df_temp.drop(columns=['position', 'duration_minutes', 'explicit', 'month'], inplace=True)
    df_temp['genre']=f'{genre}'
    df_all_genres = pd.concat([df_all_genres, df_temp])
```
